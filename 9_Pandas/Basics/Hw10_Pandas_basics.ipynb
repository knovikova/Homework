## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power
```

```python
# выбираю страны Балтии:
Baltic = power[(power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia')]

# далее из них осталяю только категории 4, 12, 21:
Baltic_categories = Baltic[(Baltic['category']==4) | (Baltic['category']==12) | (Baltic['category']==21)]

# за период с 2005 по 2010 года:
Baltic_categories_years = Baltic_categories[(Baltic_categories['year'] >= 2005) & (Baltic_categories['year'] <= 2010)]

# суммарное потребление (без учета отрицательных значений quantity): 
total = Baltic_categories_years.query('quantity > 0')['quantity'].sum()
total
```
