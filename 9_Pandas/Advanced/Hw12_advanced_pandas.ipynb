## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
# переводим timestamp из unixtime в формат даты
ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings.head(21)
```

```python
# группируем по user id, расчитываем максимальную и минимальную дату и количество оценок, поставленных каждым пользователем
ratings_grouped = ratings.groupby('userId').agg({'date': ['min', 'max'], 'movieId': 'count'})
ratings_grouped
```

```python
# отбираем пользователей, которые поставили 100 и более оценок
exceed_100 = ratings_grouped.loc[ratings_grouped.movieId['count'] >= 100]
exceed_100
```

```python
# для оставшихся пользователей рассчитываем среднее время жизни
exceed_100.date['max'] - exceed_100.date['min']
```

## Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
 - таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
 - аналогичную таблицу по типам выручки с указанием адреса клиента

Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки

auto - автомобильные перевозки

air - воздушные перевозки

client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# вертикальная конкатенация трех таблиц
transportation = pd.concat([rzd, auto, air])
transportation
```

```python
# таблица с тремя типами выручки для каждого client_id без указания адреса клиента
transportation_grouped = transportation.groupby('client_id').sum()
transportation_grouped
```

```python
# таблиц по типам выручки с указанием адреса клиента. использовала outer, чтобы выявить клиентов, которые есть в базе, но не услугами не пользуются
transportation_grouped.merge(client_base, how='outer', on='client_id')
```
