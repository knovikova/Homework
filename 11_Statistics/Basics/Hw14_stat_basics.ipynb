## Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# рассчет межквартильного размаха - 4_rectal_t

q1_4 = sample_quantit['4_rectal_t'].quantile(0.25)
q3_4 = sample_quantit['4_rectal_t'].quantile(0.75)
iqr_4 = q3_4 - q1_4
lower_bound_4 = q1_4 - (1.5*iqr_4)
higher_bound_4 = q3_4 + (1.5*iqr_4)
lower_outliners_4 = sample_quantit.loc[sample_quantit['4_rectal_t'].lt(lower_bound_4)]
higher_outliners_4 = sample_quantit.loc[sample_quantit['4_rectal_t'].gt(higher_bound_4)]
outliners_4 = pd.concat([lower_outliners_4, higher_outliners_4])
outliners_4
```
